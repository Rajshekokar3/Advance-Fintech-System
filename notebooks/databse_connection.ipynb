```python
from sqlalchemy import create_engine

DB_USER = "postgres"
DB_PASSWORD = '7984'
DB_HOST = "localhost"
DB_PORT = "5432"
DB_NAME = "Fintech"

engine = create_engine(
    f"postgresql+psycopg2://{DB_USER}:{DB_PASSWORD}@{DB_HOST}:{DB_PORT}/{DB_NAME}"
)
```

```python
import pandas as pd
nse_data = pd.read_csv(
    "C:/Users/hp r5/OneDrive/Desktop/Fintect System/Advance-Fintech-System/data/nse_data/Quote-Equity-RELIANCE-EQ-17-01-2025-17-01-2026.csv",  encoding="utf-8"
)
```

```python
nse_data.rename(columns={
    "DATE": "trade_date",
    "SERIES": "series",
    "OPEN": "open_price",
    "HIGH": "high_price",
    "LOW": "low_price",
    "PREV. CLOSE": "prev_close",
    "LTP": "ltp",
    "CLOSE": "close_price",
    "VWAP": "vwap",
    "52W H": "high_52w",
    "52W L": "low_52w",
    "VOLUME": "volume",
    "VALUE": "value",
    "NO. OF  TRADES": "no_of_trades"
}, inplace=True)
```

```python
numeric_cols = [
    "open_price", "high_price", "low_price",
    "prev_close", "ltp", "close_price",
    "vwap", "high_52w", "low_52w",
    "volume", "value", "no_of_trades"
]
```

```python
for col in numeric_cols:
    nse_data[col] = (
        nse_data[col]
        .astype(str)
        .str.replace(",", "", regex=False)
        .replace("nan", None)
        .astype(float)
    )
```

```python
nse_data.to_sql(
    name="reliance_data",
    con=engine,
    if_exists="append",   # use "replace" only once if needed
    index=False,
    method="multi"
)

print("✅ Data inserted successfully")
```
